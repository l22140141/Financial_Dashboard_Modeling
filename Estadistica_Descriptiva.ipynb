#Estadística descriptiva

```r
install.packages("readxl")
library(readxl)
```

```r
datos <- read_excel("data.xlsx")
head(datos)
```

#Medidas de tendencia central

```r
mean(datos$Estatura)
```

```r
apply(as.matrix(datos[,2:4]), 2,mean)
```

```r
lapply(datos[,2:4], mean)
```

```r
unlist(lapply(datos[,2:4], mean)) # unlist
```

```r
median(datos$Estatura)
```

```r
apply(as.matrix(datos[,2:4]),2, median)
```

```r
unlist(lapply(datos[,2:4], median)) # unlist
```

#Medidas de disperción

```r
var(datos$Estatura)
```

```r
sd(datos$peso)
```

```r
apply(as.matrix(datos[,2:4]), 2, sd)
```

```r
unlist(lapply(datos[,2:4], sd)) # unlist
```

#Medidas de localización

```r
quantile(datos$Estatura)
```
